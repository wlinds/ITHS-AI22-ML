**1. Vad är data leakage, varför är det dåligt och hur kan du undvika data leakage?**

Dataleakage är när train och test data inte separerats korrekt och någonstans i processen används samma data till träning som till testning, vilket leder till overfit eller overperformance, som i sin tur leder till att modellen inte performar väl på ny data.

**2. Beskriv vad poängen kan vara att dela upp datasetet till träningsdata och testdata för maskininlärning?**

För att testa en modell åsdiolägger vi en mindre proportion (10-30% ish) som inte används vid träning. Om vi inte hade delat upp datan på detta vis hade vi behövt hämta ny data för att testa modellen.

**3. Beskriv vad poängen kan vara att dela upp datasetet till tränings-, validerings- och testdataset för maskininlärning?**

När vi tränar en modell och anpassar hyperparametrar kan modellen bli overfitted, alltså väldigt bra på just träningsdatan, men presterar sedan mycket sämre på testdatan. Med valideringsdatan ser vi till att unvika en overfit modell.

**4. Hur skiljer sig batch gradient descent och mini batch gradient descent?**

Batch GD uppdaterar hyperparametrar efter att ha intererat genom all träningsdata. Mini-batch GD delar upp datan i flera mindre chunks (mini-batches) och uppdaterar parametrar efter varje mini-batch, alltså flera gånger per fullständig iteration av all träningsdata.

Batch GD kan fastna lokalt medan mini-batch GD kan röra sig mer fritt.
