````python
import numpy as np
seed = 42 # for reproducibility and reference
np.random.seed(seed)

# x = minutes in call / month
# y = total SEK / month

np.random.seed(seed)

# simulate 400 x-values from the random variable X~N(100,100)
x = abs(np.random.normal(loc=100, scale=100, size=400))

# simulate 400 noise values from the random variable epsilon ~N(0,50)
noise = np.random.normal(loc=0, scale=50, size=400)

y = 2*x + 25 + noise

import seaborn as sns
sns.set_style('darkgrid')
a = sns.regplot(x=x, y=y)
a.set(title = "a)")
a.set(xlabel="total minutes in call / month")
a.set(ylabel="total SEK / month")
````

````python
print(len(x), len(y))

# if no one talks more than 300 minutes / month
x_outliers = x[x >= 300]

# negative cost not possible
y_outliers = y[y < 0]

print(len(x_outliers), len(y_outliers))

# using masking to remove outliers
mask_x = np.isin(x, x_outliers)
mask_y = np.isin(y, y_outliers)

# tilde ~ operator to negate the mask
new_x = x[~mask_x]
new_y = y[~mask_y]

print(len(new_x), len(new_y))

# no idea if this will work, but imma try TODO: <-- This might have caused some issue, check up
fill_x = np.full(len(x) - len(new_x), np.nan)
fill_y = np.full(len(y) - len(new_y), np.nan)


x2 = np.concatenate((new_x, fill_x))
y2 = np.concatenate((new_y, fill_y))

print(len(x2), len(y2))

b = sns.regplot(x=x2, y=y2)
b.set(title = "b)")
b.set(xlabel="total minutes in call / month")
b.set(ylabel="total SEK / month")
````

````python
# c) add vals to df
import pandas as pd

df = pd.DataFrame({'x': x2, 'y': y2})

# not sure if the fill np.nan degraded the stats, need to check that up
df.describe().T
````

````python
from __future__ import annotations
# from sklearn.model_selection import train_test_split -> without this?


#df = pd.DataFrame({'x': x2, 'y': y2})

#train_fraction = int(len(df)*.7)

# ugly af but works
def train_test_split(X: pd.DataFrame, y: pd.DataFrame, train_fraction=.7, random_state=42, replace=False) -> tuple:

    df_X, df_y = X, y

    X_train = df_X.sample(frac=train_fraction, random_state = random_state, replace=False)
    y_train = df_y.sample(frac=train_fraction, random_state = random_state, replace=False)

    X_test = df_X.drop(X_train.index)
    y_test = df_y.drop(y_train.index)

    return X_train, X_test, y_train, y_test
````
